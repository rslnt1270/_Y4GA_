----------------->  r"\n([^\n]+)\n\+\d+" <-------- ExpresiÃ³n Regular
Busca un ladrillo que tenga exactamente estas caracterÃ­sticas

Primero; r= En Python (y en casi todos los lenguajes de programaciÃ³n), la barra invertida \ es un carÃ¡cter de escape. 

r: Le dice a Python: "No interpretes estas "\".

#los datos son solo una larga cadena de caracteres. El salto de lÃ­nea es la Ãºnica seÃ±al. "AquÃ­ termina una idea y comienza otra".


-Busca el final de la lÃ­nea anterior (\n).
-Captura todo lo que sigue mientras no sea otro salto de lÃ­nea ([^\n]+).
-Detente cuando encuentres el final de este renglÃ³n (\n).

Ejemplo: 
Texto crudo:
****  RESUMEN SEMANAL\nRoman Yair Ortega\n+5212345678  *******

\n	Detecta el espacio despuÃ©s de "SEMANAL".
([^\n]+) Empieza a leer "Roman..." y se detiene justo antes del siguiente espacio.

\n: Confirma que ahÃ­ terminÃ³ el nombre y no hay mÃ¡s palabras en ese renglÃ³n.


Los ParÃ©ntesis ( ) : El "Contenedor"
Decirle a Python: "De todo el patrÃ³n largo que estÃ¡s buscando, esto es lo Ãºnico que me interesa extraer y guardar en una variable".

[^\n]: Significa "busca cualquier carÃ¡cter (letras, espacios, nÃºmeros, puntos, emojis) siempre y cuando NO sea un salto de lÃ­nea".

Los Corchetes con Gorrito [^ ] : La "NegaciÃ³n"
Un ^ justo despuÃ©s del primer corchete, significa "Cualquier cosa MENOS esto".

([^\n]+) El SÃ­mbolo de MÃ¡s "+". 
Es un cuantificador. Le dice a la bÃºsqueda: "No te detengas en la primera letra; sigue buscando y atrapando caracteres mientras se cumpla la condiciÃ³n anterior".


"Empieza a guardar todo lo que leas aquÃ­... sigue leyendo... Â¿una 'R'?, guÃ¡rdala... Â¿un espacio?, guÃ¡rdalo... Â¿la 'O'?, guÃ¡rdala... Â¡CUIDADO!, ahÃ­ viene un \n (salto de lÃ­nea), Â¡DETENTE! y cierra el grupo aquÃ­".


VERSIÃ“N SENCILLA EXPRESION REGULAR

name_pattern = r"\n([a-zA-ZÃ¡Ã©Ã­Ã³ÃºÃÃ‰ÃÃ“ÃšÃ±Ã‘ ]+)\n"

a-z: Busca cualquier letra minÃºscula de la 'a' a la 'z'.

A-Z: Busca cualquier letra mayÃºscula de la 'A' a la 'Z'.

Ã¡Ã©Ã­Ã³ÃºÃÃ‰ÃÃ“ÃšÃ±Ã‘: Como Uber opera en espaÃ±ol, debemos incluir explÃ­citamente las tildes y la 'Ã±' para que no se detenga si tu nombre es "AdriÃ¡n" o "PeÃ±a".

(   ) (un espacio fÃ­sico):Si no se coloca el espacio dentro de los corchetes, la bÃºsqueda se detendrÃ¡. 

[a-zA-Z ] <------  "Solo acepta letras y espacios".


"Solo saca de la canasta las manzanas y las peras". 
(Nombrar cada cosa: a-z, A-Z, Ã¡-Ã©, etc.).


"Saca todo lo que haya en la canasta, menos las piedras".

La versiÃ³n ([^\n]+) no sabe quÃ© es una "A" o una "z". Lo Ãºnico que sabe es que el carÃ¡cter \n (salto de lÃ­nea) es su Ãºnica prohibiciÃ³n.

Dentro de los corchetes [], el sÃ­mbolo ^ actÃºa como un interruptor de "invertir selecciÃ³n".

[\n] -> Significa: "Busca solo saltos de lÃ­nea".

[^\n] -> Significa: "Busca cualquier cosa en el universo (letras chinas, emojis, espacios, nÃºmeros, tildes, sÃ­mbolos), excepto un salto de lÃ­nea".


El signo + le dice a la Regex: "No te conformes con el primer carÃ¡cter que encuentres. Sigue 'comiÃ©ndote' caracteres uno tras otro mientras sigan no siendo un salto de lÃ­nea".

---------------> r"(\d{2}/\d{2}/\d{4})" <-------------

r : Asegura que las barras invertidas \ lleguen directamente al motor de bÃºsqueda (import re)
sin que Python las toque.
( ... ): Encierra la fecha para que podamos extraerla fÃ¡cilmente despuÃ©s con .group().
\d : Es el comando para buscar cualquier nÃºmero del 0 al 9.
{n}: Indica cuÃ¡ntas veces debe repetirse el carÃ¡cter anterior:

        \d{2}: Busca exactamente dos nÃºmeros seguidos (ej. 15).

        \d{4}: Busca exactamente cuatro nÃºmeros seguidos (ej. 2025).

/ (Diagonal fecha): Las diagonales no tienen un significado especial en Regex, asÃ­ que se buscan tal cual aparecen en el texto.


En el texto de Uber aparece lo siguiente:
"... Resumen semanal 15/12/2025 4 a.m. - 22/12/2025 ..."

El escÃ¡ner avanza: Lee "Resumen semanal" y lo ignora porque no son nÃºmeros.
Encuentra el primer nÃºmero: Ve el 1. Revisa si hay otro nÃºmero despuÃ©s (5). 
Verifica el separador: Â¿Lo que sigue es una /? 
Busca otros dos nÃºmeros, otra diagonal y finalmente cuatro nÃºmeros.
Al cumplirse toda la "receta", captura 15/12/2025.

Imagina que caminas por una playa llena de basura (el texto crudo).

La Regex del nombre era como buscar "cualquier objeto que no sea arena".
La Regex de la fecha es un detector de metales configurado para sonar solo si encuentra algo que mide exactamente 2 cm, luego una barra, 2 cm, otra barra y 4 cm.

```python
# Regex para Total de Viajes 
# Busca la frase y atrapa los nÃºmeros que le siguen
trips_pattern = r"Cantidad total de viajes completados[^\d]*(\d+)"
match_trips = re.search(trips_pattern, initial_text)
total_trips = int(match_trips.group(1)) if match_trips else 0

# Regex para Tiempo Conectado
time_pattern = r"Tiempo efectivo trabajado.*?(\d+)\s*h\s*(\d+)\s*m"
match_time = re.search(time_pattern, initial_text, re.DOTALL)
if match_time:
    hours = int(match_time.group(1))
    minutes = int(match_time.group(2))
    total_hours_decimal = round(hours + (minutes / 60.0), 2)  #<----Convertimos a decimal para futuros cÃ¡lculos
    formatted_time = f"{hours}h {minutes}m"
else:
    formatted_time = "Not found"

# Regex para Propinas ------

tips_pattern = r"Propinas otorgadas por usuarios[^\d\$]*\$([\d,\.]+)"
match_tips = re.search(tips_pattern, initial_text, re.IGNORECASE)

if match_tips:
    tips_str = match_tips.group(1).replace(",", "")
    tips_amount = float(tips_str)
    formatted_tips = f"${tips_amount:,.2f}"
else:
    tips_amount = 0.0
    formatted_tips = "$0.00"

# Regex para Incentivos
incentives_pattern = r"ComisiÃ³n por promociÃ³n y fomento[^\d\$]*\$([\d,\.]+)"
match_incentives = re.search(incentives_pattern, initial_text, re.IGNORECASE)

if match_incentives:
    incentives_str = match_incentives.group(1).replace(",", "")
    incentives_amount = float(incentives_str)
    formatted_incentives = f"${incentives_amount:,.2f}"
else:
    incentives_amount = 0.0
    formatted_incentives = "$0.00"

# Regex para Impuestos

taxes_pattern = r"Impuestos[^\d\-]*-\$([\d,\.]+)"
match_taxes = re.search(taxes_pattern, initial_text, re.IGNORECASE)

if match_taxes:
    taxes_str = match_taxes.group(1).replace(",", "")
    taxes_amount = -float(taxes_str)
    formatted_taxes = f"-${abs(taxes_amount):,.2f}"
else:
    taxes_amount = 0.0
    formatted_taxes = "$0.00"

# Regex para Monto bruto
gross_pattern = r"Monto bruto generado[^\d]*([\d,]+\.\d{2})"
match_gross = re.search(gross_pattern, text_page1)

if match_gross:
    gross_str = match_gross.group(1)
    # Remove the comma to convert the string into a mathematical float
    gross_float = float(gross_str.replace(",", ""))
    # Format it back to a readable currency string
    formatted_gross_amount = f"${gross_float:,.2f}"
else:
    formatted_gross_amount = "Not found"

print("-" * 50)
print("RESUMEN SEMANAL UBER ðŸ“Š")
print("-" * 50)
print(f" ðŸ‘¤ CONDUCTOR:             {driver_name}")
print(f" ðŸ“… SEMANA:                {weekly_period}")
print(f" ðŸš— VIAJES COMPLETADOS:    {total_trips}")
print(f" â±ï¸ TIEMPO CONECTADO:      {formatted_time}")
print("-" * 50)
print("DESGLOSE FINANCIERO")
print(f" ðŸ’µ MONTO BRUTO GENERADO:  {formatted_gross_amount}")
print(f" ðŸª™ PROPINAS:              {formatted_tips}")
print(f" ðŸŽ INCENTIVOS:            {formatted_incentives}")
print(f" ðŸ’¸ IMPUESTOS:             {formatted_taxes}")
print("-" * 50)
```
