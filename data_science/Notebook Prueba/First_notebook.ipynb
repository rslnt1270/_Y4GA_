----------------->  r"\n([^\n]+)\n\+\d+" <-------- Expresión Regular
Busca un ladrillo que tenga exactamente estas características

Primero; r= En Python (y en casi todos los lenguajes de programación), la barra invertida \ es un carácter de escape. 

r: Le dice a Python: "No interpretes estas "\".

#los datos son solo una larga cadena de caracteres. El salto de línea es la única señal. "Aquí termina una idea y comienza otra".


-Busca el final de la línea anterior (\n).
-Captura todo lo que sigue mientras no sea otro salto de línea ([^\n]+).
-Detente cuando encuentres el final de este renglón (\n).

Ejemplo: 
Texto crudo:
****  RESUMEN SEMANAL\nRoman Yair Ortega\n+5212345678  *******

\n	Detecta el espacio después de "SEMANAL".
([^\n]+) Empieza a leer "Roman..." y se detiene justo antes del siguiente espacio.

\n: Confirma que ahí terminó el nombre y no hay más palabras en ese renglón.


Los Paréntesis ( ) : El "Contenedor"
Decirle a Python: "De todo el patrón largo que estás buscando, esto es lo único que me interesa extraer y guardar en una variable".

[^\n]: Significa "busca cualquier carácter (letras, espacios, números, puntos, emojis) siempre y cuando NO sea un salto de línea".

Los Corchetes con Gorrito [^ ] : La "Negación"
Un ^ justo después del primer corchete, significa "Cualquier cosa MENOS esto".

([^\n]+) El Símbolo de Más "+". 
Es un cuantificador. Le dice a la búsqueda: "No te detengas en la primera letra; sigue buscando y atrapando caracteres mientras se cumpla la condición anterior".


"Empieza a guardar todo lo que leas aquí... sigue leyendo... ¿una 'R'?, guárdala... ¿un espacio?, guárdalo... ¿la 'O'?, guárdala... ¡CUIDADO!, ahí viene un \n (salto de línea), ¡DETENTE! y cierra el grupo aquí".


VERSIÓN SENCILLA EXPRESION REGULAR

name_pattern = r"\n([a-zA-ZáéíóúÁÉÍÓÚñÑ ]+)\n"

a-z: Busca cualquier letra minúscula de la 'a' a la 'z'.

A-Z: Busca cualquier letra mayúscula de la 'A' a la 'Z'.

áéíóúÁÉÍÓÚñÑ: Como Uber opera en español, debemos incluir explícitamente las tildes y la 'ñ' para que no se detenga si tu nombre es "Adrián" o "Peña".

(   ) (un espacio físico):Si no se coloca el espacio dentro de los corchetes, la búsqueda se detendrá. 

[a-zA-Z ] <------  "Solo acepta letras y espacios".


"Solo saca de la canasta las manzanas y las peras". 
(Nombrar cada cosa: a-z, A-Z, á-é, etc.).


"Saca todo lo que haya en la canasta, menos las piedras".

La versión ([^\n]+) no sabe qué es una "A" o una "z". Lo único que sabe es que el carácter \n (salto de línea) es su única prohibición.

Dentro de los corchetes [], el símbolo ^ actúa como un interruptor de "invertir selección".

[\n] -> Significa: "Busca solo saltos de línea".

[^\n] -> Significa: "Busca cualquier cosa en el universo (letras chinas, emojis, espacios, números, tildes, símbolos), excepto un salto de línea".


El signo + le dice a la Regex: "No te conformes con el primer carácter que encuentres. Sigue 'comiéndote' caracteres uno tras otro mientras sigan no siendo un salto de línea".

---------------> r"(\d{2}/\d{2}/\d{4})" <-------------

r : Asegura que las barras invertidas \ lleguen directamente al motor de búsqueda (import re)
sin que Python las toque.
( ... ): Encierra la fecha para que podamos extraerla fácilmente después con .group().
\d : Es el comando para buscar cualquier número del 0 al 9.
{n}: Indica cuántas veces debe repetirse el carácter anterior:

        \d{2}: Busca exactamente dos números seguidos (ej. 15).

        \d{4}: Busca exactamente cuatro números seguidos (ej. 2025).

/ (Diagonal fecha): Las diagonales no tienen un significado especial en Regex, así que se buscan tal cual aparecen en el texto.


En el texto de Uber aparece lo siguiente:
"... Resumen semanal 15/12/2025 4 a.m. - 22/12/2025 ..."

El escáner avanza: Lee "Resumen semanal" y lo ignora porque no son números.
Encuentra el primer número: Ve el 1. Revisa si hay otro número después (5). 
Verifica el separador: ¿Lo que sigue es una /? 
Busca otros dos números, otra diagonal y finalmente cuatro números.
Al cumplirse toda la "receta", captura 15/12/2025.

Imagina que caminas por una playa llena de basura (el texto crudo).

La Regex del nombre era como buscar "cualquier objeto que no sea arena".
La Regex de la fecha es un detector de metales configurado para sonar solo si encuentra algo que mide exactamente 2 cm, luego una barra, 2 cm, otra barra y 4 cm.

```python
# Regex para Total de Viajes 
# Busca la frase y atrapa los números que le siguen
trips_pattern = r"Cantidad total de viajes completados[^\d]*(\d+)"
match_trips = re.search(trips_pattern, initial_text)
total_trips = int(match_trips.group(1)) if match_trips else 0

# Regex para Tiempo Conectado
time_pattern = r"Tiempo efectivo trabajado.*?(\d+)\s*h\s*(\d+)\s*m"
match_time = re.search(time_pattern, initial_text, re.DOTALL)
if match_time:
    hours = int(match_time.group(1))
    minutes = int(match_time.group(2))
    total_hours_decimal = round(hours + (minutes / 60.0), 2)  #<----Convertimos a decimal para futuros cálculos
    formatted_time = f"{hours}h {minutes}m"
else:
    formatted_time = "Not found"

# Regex para Propinas ------

tips_pattern = r"Propinas otorgadas por usuarios[^\d\$]*\$([\d,\.]+)"
match_tips = re.search(tips_pattern, initial_text, re.IGNORECASE)

if match_tips:
    tips_str = match_tips.group(1).replace(",", "")
    tips_amount = float(tips_str)
    formatted_tips = f"${tips_amount:,.2f}"
else:
    tips_amount = 0.0
    formatted_tips = "$0.00"

# Regex para Incentivos
incentives_pattern = r"Comisión por promoción y fomento[^\d\$]*\$([\d,\.]+)"
match_incentives = re.search(incentives_pattern, initial_text, re.IGNORECASE)

if match_incentives:
    incentives_str = match_incentives.group(1).replace(",", "")
    incentives_amount = float(incentives_str)
    formatted_incentives = f"${incentives_amount:,.2f}"
else:
    incentives_amount = 0.0
    formatted_incentives = "$0.00"

# Regex para Impuestos

taxes_pattern = r"Impuestos[^\d\-]*-\$([\d,\.]+)"
match_taxes = re.search(taxes_pattern, initial_text, re.IGNORECASE)

if match_taxes:
    taxes_str = match_taxes.group(1).replace(",", "")
    taxes_amount = -float(taxes_str)
    formatted_taxes = f"-${abs(taxes_amount):,.2f}"
else:
    taxes_amount = 0.0
    formatted_taxes = "$0.00"

# Regex para Monto bruto
gross_pattern = r"Monto bruto generado[^\d]*([\d,]+\.\d{2})"
match_gross = re.search(gross_pattern, text_page1)

if match_gross:
    gross_str = match_gross.group(1)
    # Remove the comma to convert the string into a mathematical float
    gross_float = float(gross_str.replace(",", ""))
    # Format it back to a readable currency string
    formatted_gross_amount = f"${gross_float:,.2f}"
else:
    formatted_gross_amount = "Not found"

print("-" * 50)
print("RESUMEN SEMANAL UBER 📊")
print("-" * 50)
print(f" 👤 CONDUCTOR:             {driver_name}")
print(f" 📅 SEMANA:                {weekly_period}")
print(f" 🚗 VIAJES COMPLETADOS:    {total_trips}")
print(f" ⏱️ TIEMPO CONECTADO:      {formatted_time}")
print("-" * 50)
print("DESGLOSE FINANCIERO")
print(f" 💵 MONTO BRUTO GENERADO:  {formatted_gross_amount}")
print(f" 🪙 PROPINAS:              {formatted_tips}")
print(f" 🎁 INCENTIVOS:            {formatted_incentives}")
print(f" 💸 IMPUESTOS:             {formatted_taxes}")
print("-" * 50)
```

CARGA MASIVA DE PDF

```python
import pdfplumber
import re
import glob
import pandas as pd

# --- CONFIGURACIÓN DE RUTAS Y PATRONES ---
RUTA_CARPETA = "pdf_uber/*.pdf"
# Patrón para extraer el rango de fechas
DATE_PATTERN = re.compile(r"(\d{1,2}\s+[a-z]{3}\s+\d{4}.*?-.*?\d{4})", re.IGNORECASE)
# Patrón para identificar y remover la hora (ej: "4 a. m.")
HORA_PATTERN = re.compile(r"\s+\d{1,2}\s+[ap]\.\s+m\.", re.IGNORECASE)
# Patrón para localizar el monto exacto en la página 3
MONTO_PATTERN = re.compile(r"Monto bruto generado\s+([\d\.,]+)")

def extraer_datos_profesional(ruta_archivo):
    """Extrae, limpia y formatea los datos de un PDF de Uber."""
    try:
        with pdfplumber.open(ruta_archivo) as pdf:
            # Foco en página 3 (Comprobante de pago)
            pagina = pdf.pages[2] if len(pdf.pages) >= 3 else pdf.pages[-1]
            texto = " ".join(pagina.extract_text().split())
            
            # 1. LIMPIEZA DE FECHA (Sin horas)
            match_date = DATE_PATTERN.search(texto)
            fechas_final = "No encontrada"
            if match_date:
                # Removemos las horas para dejar solo: dia mes año - dia mes año
                fechas_final = HORA_PATTERN.sub("", match_date.group(1)).strip()
            
            # 2. LIMPIEZA DE MONTO (Formato contable $)
            match_monto = MONTO_PATTERN.search(texto)
            monto_formateado = "$0.00"
            if match_monto:
                # Conversión técnica de formato latino a float de Python
                valor_num = float(match_monto.group(1).replace(".", "").replace(",", "."))
                # Formateo con comas de miles y signo de pesos
                monto_formateado = "${:,.2f}".format(valor_num)

            return {
                "Conductor": "Roman Yair Ortega",
                "Fechas": fechas_final,
                "Monto_Bruto": monto_formateado
            }
    except Exception as e:
        print(f"❌ Error en {ruta_archivo}: {e}")
        return None

# --- EJECUCIÓN DEL PROCESO ---
archivos_pdf = glob.glob(RUTA_CARPETA)
todos_los_datos = []

print(f"🔍 Procesando {len(archivos_pdf)} archivos con formato mejorado...\n")

for archivo in archivos_pdf:
    resultado = extraer_datos_profesional(archivo)
    if resultado:
        todos_los_datos.append(resultado)
        print(f"✅ Limpio: {resultado['Fechas']} | {resultado['Monto_Bruto']}")

# 3. CREACIÓN DE TABLA FINAL
df_final = pd.DataFrame(todos_los_datos)

# Eliminamos cualquier rastro de la columna 'Archivo' si existiera y mostramos
print("\n🚀 Fase 1 finalizada. Estructura profesional alcanzada:")
display(df_final)
```
